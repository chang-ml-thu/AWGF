```python
from __future__ import division, print_function
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from dynamics import Dynamics
import time as tm
import os

rng = np.random.RandomState()
np.random.seed(44)
```

```python
def w1(z):
    return tf.sin(2.*np.pi*z[0]/4.)
def w2(z):
    return 3.*tf.exp(-.5*(((z[0]-1.)/.6))**2)
def w3(z):
    return 3.*(1+tf.exp(-(z[0]-1.)/.3))**-1
def pot0(z):
    z = tf.transpose(z)
    return .5*((tf.norm(z, ord=2, axis=0)-0.)/.4)**2 - tf.log(tf.exp(-.5*((z[0]-0.)/.8)**2) )
def pot1(z):
    z = tf.transpose(z)
    return .5*((tf.norm(z, ord=2, axis=0)-3.)/.5)**2 - tf.log(tf.exp(-.5*((z[0]-3.)/.5)**2) + tf.exp(-.5*((z[0]+3.)/.5)**2))
def pot2(z):
    z = tf.transpose(z)
    return .5*((z[1]-w1(z))/.4)**2
def pot3(z):
    z = tf.transpose(z)
    return -tf.log(tf.exp(-.5*((z[1]-w1(z))/.35)**2) + tf.exp(-.5*((z[1]-w1(z)+w2(z))/.35)**2))
def pot4(z):
    z = tf.transpose(z)
    return -tf.log(tf.exp(-.5*((z[1]-w1(z))/.4)**2) + tf.exp(-.5*((z[1]-w1(z)+w3(z))/.35)**2))
```

# Global Parameters

```python
num_particles = 200
num_latent = 2
num_iter = 400
num_fig = 1
savefig = False
plt_range_limit = [-4,4]
plt_num_points = 50
get_logp = lambda z: -pot1(z)
logp_name = 'pot1'
```

```python
samples0 = np.random.randn(num_particles, num_latent)
samples_tf = tf.Variable(samples0)

grid_side = np.linspace(plt_range_limit[0], plt_range_limit[1], plt_num_points)
mesh_z1, mesh_z2 = np.meshgrid(grid_side, grid_side)
zv = np.hstack([mesh_z1.reshape(-1, 1), mesh_z2.reshape(-1, 1)])
zp_tf  = tf.placeholder(tf.float32, [None, num_latent])
prob_tf = tf.exp(get_logp(zp_tf))
with tf.Session() as sess:
    phat_z = sess.run(prob_tf, feed_dict={zp_tf: zv}) 
phat_z = phat_z.reshape([plt_num_points, plt_num_points])
z_min, z_max = -np.abs(phat_z).max(), np.abs(phat_z).max()

def showfig_smp(pm):
    op_samples, dninfo = Dynamics(pm.dnType, pm).evolve(samples_tf, get_logp=get_logp)
    fig, ax = plt.subplots(1, num_fig, sharey=True, sharex=True, figsize=(6,6))#(4*num_fig,4))
    if num_fig == 1:
        figrange = [0]; ax = [ax]
    else:
        figrange = range(1, num_fig)
        ax[0].set_xlim(plt_range_limit); ax[0].set_ylim(plt_range_limit);
        ax[0].set_aspect('equal'); ax[0].set_xticks([]); ax[0].set_yticks([])
        # ax[0].pcolormesh(mesh_z1, mesh_z2, phat_z, cmap='RdBu', vmin=z_min, vmax=z_max)
        ax[0].contourf(mesh_z1, mesh_z2, phat_z, cmap='Blues', vmin=0, vmax=z_max*1.8)
        ax[0].scatter(samples0[:,0], samples0[:,1], c="dimgrey", marker="o", s=30, linewidths=0, edgecolor='white', alpha=1)
    with tf.Session() as sess:
        tf.global_variables_initializer().run()
        for i in figrange:
            for j in range(num_iter-1): sess.run(op_samples)
            samples = sess.run([op_samples, dninfo.L_samples])[1][0]
            ax[i].set_aspect('equal'); ax[i].set_xticks([]); ax[i].set_yticks([])
            #ax[i].pcolormesh(mesh_z1, mesh_z2, phat_z, cmap='RdBu', vmin=z_min, vmax=z_max)
            ax[i].contourf(mesh_z1, mesh_z2, phat_z, cmap='Blues', vmin=0, vmax=z_max*1.8)
            ax[i].scatter(samples[:,0], samples[:,1], c='dimgrey', marker="o", s=30, linewidths=0, edgecolor='white', alpha=1)
    if savefig:
        dirname = './synthetic_res_' + logp_name + '/'
        if not os.path.isdir(dirname): os.makedirs(dirname)
        fig.savefig( dirname \
            + '_'.join([pm.dnType, pm.accType, pm.bwType if hasattr(pm, 'bwType') else 'void', pm.optType]) \
            + '.pdf', bbox_inches='tight')
    return fig, ax
```

# SVGD

```python
class PM:
    dnType = 'SVGD'; dnNormalize = False
    accType = 'wgd'
    optType = 'gd'; stepsize = 3e-1
    bwType = 'med'
t0 = tm.time(); fig, ax = showfig_smp(PM()); t1 = tm.time(); print(t1-t0)
```
